```python
from functions import *

# load orbit data
Juice = get_spice_data('juice', 'callisto', 'cphio')
Galileo, Galileo_meas = get_pds_data()

# save CA data
Juice_CA = closest_approach_data(Juice)
Galileo_CA = closest_approach_data(Galileo)
```

```python
# Plotting trajectories

lim = 30

def add_callisto_circle(ax):
    callisto = plt.Circle((0, 0), 1, color='black')
    ax.add_patch(callisto)
    return ax

plt.style.use('ggplot')
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# Defining and applying the common limits
common_xlim = (-lim, lim)  
common_ylim = (-lim, lim) 
common_zlim = (-lim, lim) 

axes[0].set_xlim(common_xlim)
axes[0].set_ylim(common_ylim)

axes[1].set_xlim(common_xlim)
axes[1].set_ylim(common_zlim)

axes[2].set_xlim(common_ylim)
axes[2].set_ylim(common_zlim)

for i in range(3):
    axes[i] = add_callisto_circle(axes[i])
    
# Set the labels
axes[2].set_xlabel('x [$R_C$]')
axes[2].set_ylabel('y [$R_C$]')

axes[1].set_xlabel('x [$R_C$]')
axes[1].set_ylabel('z [$R_C$]')

axes[0].set_xlabel('y [$R_C$]')
axes[0].set_ylabel('z [$R_C$]')

# Create a color palette with 21 different colors
from itertools import cycle

colors = plt.cm.tab20.colors
color_cycle = cycle(colors)

for orbit, vector in Juice.items():
    color = next(color_cycle)

    axes[0].plot(vector[2] / R_C, vector[3] / R_C, color=color)
    axes[1].plot(vector[1] / R_C, vector[3] / R_C, color=color, label=orbit)
    axes[2].plot(vector[1] / R_C, vector[2] / R_C, color=color)

#axes[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=7)
plt.tight_layout()
```

```python
# Compare CA of Galileo and Juice

plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(9,5))

plt.xticks(rotation=90) 

for orbit, vector in Juice_CA.items():
    vector[1:5] = vector[1:5] / R_C
    ax.plot(orbit, vector[4], '*', markersize=10, color='#f94144')

for orbit, vector in Galileo_CA.items():
    vector[1:5] = vector[1:5] / R_C
    ax.plot(orbit, vector[4], '*', markersize=10, color='#577590')

# Create custom legend handles and labels
juice_legend = mlines.Line2D([], [], color='#f94144', marker='*', markersize=10, label='Juice')
galileo_legend = mlines.Line2D([], [], color='#577590', marker='*', markersize=10, label='Galileo')

# Add the legend with custom handles and labels
ax.legend(handles=[juice_legend, galileo_legend])
```
