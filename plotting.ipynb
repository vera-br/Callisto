```python
# 3D Plots - they don't pop out in the Jupyter Notebook, but would in a normal Python script

# Defining and applying the common limits
lim = 3

common_xlim = (-lim, lim)  
common_ylim = (-lim, lim) 
common_zlim = (-lim, lim) 

# Make Callisto sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

i = 0

for orbit, vector in Juice.items():
    color = next(color_cycle)

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    ax.set_xlim(common_xlim)
    ax.set_ylim(common_ylim)
    ax.set_zlim(common_zlim)

    ax.set_xlabel('x [$R_C$]')
    ax.set_ylabel('y [$R_C$]')
    ax.set_zlabel('z [$R_C$]')

    # Plot Callisto surface
    ax.scatter(x, y, z, color='xkcd:dull brown')

    ax.plot(vector[1] / R_C, vector[2] / R_C, vector[3] / R_C, color=color)

    i += 1
    plt.title('Orbit%s in CphiO' % (i))
    plt.show()
```
